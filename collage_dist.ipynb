```python
import torch
import torchvision
import torchvision.models as models
from torchvision import transforms
import numpy as np
from PIL import Image
from torch import nn
import matplotlib.pyplot as plt 
import copy
import cv2
import piq

def pilToNumpy(img):
    return np.array(img)

def NumpyToPil(img):
    return Image.fromarray(img)

def split_image(image, window_size):
    """
    Split an image into non-overlapping windows of the specified size.

    Parameters:
    - image: Input image (NumPy array).
    - window_size: Size of the windows (tuple of width and height).

    Returns:
    - List of windows (NumPy arrays).
    """
    

    height, width = image.shape[:2]
    window_width, window_height = window_size

    windows = []

    for y in range(0, height, window_height):
        for x in range(0, width, window_width):
            window = image[y:y + window_height, x:x + window_width]
            windows.append(window)

    return windows

def recompose_image(windows, original_size):
    """
    Recompose an image from a list of windows.

    Parameters:
    - windows: List of windows (PIL Image objects).
    - original_size: Original size of the image (tuple of width and height).

    Returns:
    - Recomposed image (PIL Image object).
    """
    width, height = original_size
    recomposed_image = Image.new('RGB', (width, height))

    window_width, window_height = NumpyToPil(windows[0]).size

    index = 0
    for y in range(0, height, window_height):
        for x in range(0, width, window_width):
            recomposed_image.paste(NumpyToPil(windows[index]), (x, y))
            index += 1

    return recomposed_image



# path_target_image = image_path = "/data/users/agomezvi/art/pirate_dog.jpeg"
path_target_image = image_path = "/data/users/agomezvi/art/totoro.png"
# path_target_image = image_path = "/data/users/agomezvi/art/loros.png"
```

Load input image, resize it

```python
input_image = cv2.imread(path_target_image)
input_image = cv2.resize(input_image, (1024, 1024), 
               interpolation = cv2.INTER_LINEAR)
input_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB)
plt.imshow(input_image)
input_image.shape
```

Split input image into non-overlapping windows

```python
window_size = (32, 32)  # Specify your desired window size
windows = split_image(input_image, window_size)

lst = [torch.from_numpy(item).float() for item in windows]
input_image_tensor = torch.stack(lst,-1).unsqueeze(0).unsqueeze(0)
input_image_tensor = torch.swapaxes(torch.swapaxes(input_image_tensor, 5, 0),4,1).squeeze()/ 255.
```

Load CIFAR100 images and save them in a tensor

```python
mean = [0.485, 0.456, 0.406]
std = [0.229, 0.224, 0.225]
batch_size = 500
transform_test = transforms.Compose([transforms.ToTensor(), transforms.Normalize(mean,std)])
testset = torchvision.datasets.CIFAR100("./",
                                        train=True,
                                        download=True,                                        
                                        transform=transform_test)
train_loader = torch.utils.data.DataLoader(
    testset, batch_size,pin_memory=True, shuffle=False, num_workers=2)

all_cifar = []
all_cifar_images = []
cont = 0

for batch in train_loader:    
   
    torch.save(batch[0], str(cont)+'_image.pt')
    cont += 1
for i in range(cont):    
   
    
    loaded = torch.load(str(i)+'_image.pt')
    all_cifar_images.append(loaded)
    
all_cifar_images_tensor = torch.vstack(all_cifar_images)
```

compare each window with all the images of CIFAR using the perceptual metric DIST, then select the CIFAR100 images most similar

```python
# for i in range(out.shape[0]):
loss = nn.MSELoss(reduction='none')
invTrans = transforms.Compose([ transforms.Normalize(mean = [ 0., 0., 0. ],
                                                     std = [ 1/std[0], 1/std[1], 1/std[2] ]),
                                transforms.Normalize(mean = [ -mean[0], -mean[1], -mean[2] ],
                                                     std = [ 1., 1., 1. ]),
                               ])


batch_size = 12500
count = 0
window_sim = []
for i in range(input_image_tensor.shape[0]):
    
    cum_loss = []
    print(i)
    
    for batch in range(all_cifar_images_tensor.shape[0]//batch_size):
        
    
        src_image = input_image_tensor[i,...].unsqueeze(0).repeat(batch_size, 1, 1, 1).cuda()  
        
        target_image = all_cifar_images_tensor[batch*batch_size:batch*batch_size+batch_size,...].cuda()
        
        cum_loss.append(piq.DISTS(reduction='none')(src_image, target_image))
    
    loss_result = torch.vstack(cum_loss)
    
    sim_imgage = torch.argmin(loss_result)
    a = torch.swapaxes(torch.swapaxes(invTrans(all_cifar_images_tensor[sim_imgage,...]), 0, 1), 1, 2).numpy()
    window_sim.append((a*256).astype(np.uint8))
```

Plot CIFAR100 photomosaic vs input image
